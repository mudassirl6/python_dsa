```python
#power set
n = [0,4,5,6]

subsets = 1 << len(n)

ans = []


for subset in range(subsets):
    temp = []
    for num in range(len(n)):
        if subset & (1<<num) != 0:
            temp.append(n[num])
            
    ans.append(temp)
    
    
ans
```

```python
def numtobin(num):
    if num == 0:
        return "00"
    
    
    
    bin = ""
    
    while num > 0:
        rem = num%2
        bin = str(rem)+bin
        num = num//2
        
    return bin

  
        
numtobin(10)
        
```

```python
def bintoNum(n):
    
    number = 0
    power = len(n)-1
    for digit in n:
        
        
        number += int(digit) *(2**power)
        power -= 1
        
    return number
        
        
bintoNum("1010")
        
        
        
```

```python
def fact(n):
    
    if n <= 1:
        return 1
    
    
    
    return n*fact(n-1)

fact(5)
```

```python
def fib(n):
    if n <= 1:
        return n
    
    return fib(n-1)+fib(n-2)
```

```python
def subsequence(i,ans,curr_arr,arr,n):
    
    if i >= n:
        if arr not in ans:
            ans.append(arr.copy())
        return
        
    arr.append(curr_arr[i])
    subsequence(i+1,ans,curr_arr,arr,n)
    arr.remove(curr_arr[i])
    subsequence(i+1,ans,curr_arr,arr,n)
    
    
curr_arr = [2,1,2,1,3]
ans = []

subsequence(0,ans,curr_arr,[],len(curr_arr))
print(ans)

# all arrs which is appended into to the ans is referring to one arr...


# The issue is with how you’re managing the arr list. In Python, lists are passed by reference, meaning any changes to arr affect all references to it. Specifically, when you do ans.append(arr), you’re appending a reference to the same list each time.

# Detailed Explanation:
# 	1.	Pass by Reference Issue:
# 	•	arr is the same object throughout all recursive calls.
# 	•	You keep modifying the same list (adding and removing elements).
# 	•	As a result, all elements in ans end up referencing the final state of arr, which is an empty list after all recursive calls complete.
# 	2.	Why Use arr.copy() or list(arr)?
# 	•	To fix this, you need to append a copy of arr at each base case.
# 	•	This way, the state of arr at that point in the recursion is preserved.
```

```python
def arrWithSum(idx,temp,arr,n,sum1,s):
    
    if idx >= n:
        if s == sum1:
            print(temp)
            return True
            
        return False
    
    temp.append(arr[idx])
    sum1 += arr[idx]
    
    if arrWithSum(idx+1,temp,arr,n,sum1,s) == True:
        return True
    
    
    temp.remove(arr[idx])
    sum1 -= arr[idx]
    
    if arrWithSum(idx+1,temp,arr,n,sum1,s) == True:
        return True
    
    
    return False
    
arr = [1,2,1]

print(arrWithSum(0,[],arr,len(arr),0,2))
```

```python
def mergesort(arr,l,r):
    if l >= r:
        return 
    
    mid = (l+r)//2
    mergesort(arr,l,mid)
    mergesort(arr,mid+1,r)
    
    merge(arr,l,mid,r)
    

def merge(arr,l,mid,r):
    
    new_arr = []
    i = l
    j = mid+1

    while i <= mid and j <= r:
        if arr[i] <= arr[j]:
            new_arr.append(arr[i])
            i += 1
        else:
            
            new_arr.append(arr[j])
            j += 1   
        
    while i <= mid:
        new_arr.append(arr[i])
        i += 1   
        
    while j <= r:
        new_arr.append(arr[j])
        j += 1
        
    for k in range(len(new_arr)):
        arr[l+k] = new_arr[k]
        
        
    

        
arr = [2,6,7,3,7,9,8]

mergesort(arr,0,len(arr)-1)

print(arr)
             
    
```

```python
def com(idx,temp,arr,ans,target,n):
    if target < 0:
        return 
    
    if idx >= n:
        if target == 0:
            ans.append(temp.copy())
            
        return 
    

    
    if arr[idx] <= target:
        temp.append(arr[idx])
        com(idx,temp,arr,ans,target-arr[idx],n)
        temp.remove(arr[idx])

        
    
    com(idx+1,temp,arr,ans,target,n)
    
    
arr = [2,3,5]
ans = []
com(0,[],arr,ans,8,len(arr))
print(ans)
        
    
```
